```python
import pandas as pd
import numpy as np
import sqlite3 as sql
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
sql_query = ('''
SELECT * FROM crashes
WHERE "Motorcycle Count" > 0
''')
conn = sql.connect('/home/michaelpfredericks/Documents/github/crash-data.db')
data = pd.read_sql_query(sql_query, conn)
conn.close()
```

```python
for i in data.columns:
    print(i)
```

```python
cols = ['Crash Record Number', 'County Name', 'Interstate', 'State Road', 'Local Road Only', 'Turnpike', 'Wet Road', 'Snow Slush Road',
        'Icy Road', 'Sudden Deer', 'Shoulder Related', 'Rear End', 'Head On / Opposite Direction Side Swipe', 'Hit Fixed Object',
        'Single Vehicle Run Off Road', 'Work Zone', 'Non Intersection', 'Intersection', 'Signalized Intersection', 'Stop Controlled Intersection', 
        'Unsignalized Intersection', 'School Bus', 'School Zone', 'Hit Deer', 'Hit Tree / Shrub', 'Hit Embankment', 'Hit Pole',
        'Hit Guide Rail', 'Hit Guide Rail End', 'Hit Barrier', 'Hit Bridge', 'Heavy Truck Related', 'Vehicle Failure', 'Train / Trolley', 
        'Alcohol Related', 'Drinking Driver', 'Underage Drinking Driver', 'Unlicensed', 'Distracted', 'Cell Phone', 'Running Red Light', 
        'Tailgating', 'Cross Median', 'Curved Road', 'Curve Driver Error', 'Limit 65 MPH', 'Speeding', 'Speeding Related', 
        'Aggressive Driving', 'Fatigue / Asleep', 'Illumination Dark', 'Running Stop Sign', 'Train', 'Trolley', 
        'Motorcycle Drinking Driver', 'Drug Related', 'Illegal Drug Related', 'Drugged Driver', 'Impaired Driver']
```

```python
df = data[cols]
```

```python
df.replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
df['MC Crash Count'] = 1
df
```

```python
df1 = pd.DataFrame(df.groupby('County Name').sum()).reset_index(level=0)
```

```python
df1
```

```python
cols
```

```python
X = df1[cols]
y = df1['MC Crash Count']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient']).sort_values('Coefficient', ascending=False)
```
