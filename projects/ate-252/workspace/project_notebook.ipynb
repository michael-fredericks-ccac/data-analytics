```python
by_county_df['Crash Record Number'].sum()
```

```python
crash_records_sum = by_county_df['Crash Record Number'].sum()
```

```python
by_county_df['% of Crashes'] = by_county_df['Crash Record Number'].apply(lambda x: x/crash_records_sum)
```

```python
by_county_df
```

```python
county_size_df = pd.read_csv('pa-county-pop-data.csv')
```

```python
county_size_df
```

```python
county_size_df.iloc[0]
```

```python
county_size_df.drop('GrowthRate', axis=1, inplace=True)
```

```python
county_size_df
```

```python
county_size_df['CTYNAME'] = county_size_df['CTYNAME'].apply(lambda x: x.split(' ')[0])
```

```python
county_size_df
```

```python
pop_sum = county_size_df['Pop'].sum()
```

```python
pop_sum
```

```python
county_size_df['%Pop'] = county_size_df['Pop'].apply(lambda x: x/pop_sum)
```

```python
county_size_df
```

```python
by_county_df.rename(columns={'Crash Record Number': 'CrashRecordCount', 'County Name': 'CTYNAME', '% of Crashes': '%Crashes'}, inplace=True)
```

```python
by_county_df.reset_index(drop=True)
by_county_df.loc[(by_county_df.CTYNAME == 'Mckean'), 'CTYNAME'] = 'McKean'
```

```python
county_data = pd.merge(by_county_df, county_size_df, on='CTYNAME')
```

```python
county_data
```

```python
county_data.head(60)
```

```python
county_data.tail(6)
```

```python
county_data_sub1 = county_data[['CTYNAME', '%Crashes']]
```

```python
county_data_sub1
```

```python
county_data_sub2 = county_data[['CTYNAME', '%Pop']]
```

```python
county_data_sub2
```

```python
county_data_sub1['Metric'] = '%Crash'
```

```python
county_data_sub1
```

```python
county_data_sub2['Metric'] = '%Pop'
```

```python
county_data_sub1.rename(columns={'%Crashes': '%'}, inplace=True)
county_data_sub2.rename(columns={'%Pop': '%'}, inplace = True)
```

```python
bypercent_county = pd.concat([county_data_sub1, county_data_sub2], ignore_index=True)
bypercent_county
```

```python
plt.figure(figsize=(15,30))
sns.set_style('darkgrid')
sns.barplot(x='%', y='CTYNAME', data=bypercent_county, hue='Metric')
plt.tick_params(labelsize=20)
```

```python
county_data
```

```python
county_data['PopCrash-Idx'] = (county_data['%Pop'] - county_data['%Crashes']) / county_data['%Pop']
```

```python
county_data
```

```python
plt.figure(figsize=(20,15))
sns.set_context('poster')
sns.barplot(x='PopCrash-Idx', y='CTYNAME', data=county_data.sort_values(by='PopCrash-Idx', ascending=False))
plt.tick_params(labelsize=10)
```
