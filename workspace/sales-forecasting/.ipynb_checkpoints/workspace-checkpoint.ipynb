```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
features = pd.read_csv('Features data set.csv')
sales = pd.read_csv('sales data-set.csv')
stores = pd.read_csv('stores data-set.csv')
```

```python
features.fillna(0, inplace=True)
```

```python
features['Markdown'] = features[['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']].sum(axis=1)
```

```python
features.drop(['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'], axis=1, inplace=True)
```

```python
df = pd.merge(sales, features, on=['Store', 'Date', 'IsHoliday'])
```

```python
df = pd.merge(df, stores, on=['Store'])
```

```python
df1 = df[(df['Store'] == 1) & (df['Dept'] == 1)]
```

```python
df1['L4W Sales'] = df1['Date'].apply(lambda x: df1[(df1['Date'] >= pd.Timestamp(x - datetime.timedelta(weeks=5))) & 
                                                   (df1['Date'] <= pd.Timestamp(x - datetime.timedelta(weeks=1)))]['Weekly_Sales'].sum())
```

```python
df1['LY Sales'] = df1['Date'].apply(lambda x: df1[df1['Date'] == pd.Timestamp(x - datetime.timedelta(weeks=52))]['Weekly_Sales'].sum())
```

```python
df1['N4W LY'] = df1['Date'].apply(lambda x: df1[(df1['Date'] >= pd.Timestamp(x - datetime.timedelta(weeks=51))) & 
                                            (df1['Date'] <= pd.Timestamp(x - datetime.timedelta(weeks=48)))]['Weekly_Sales'].sum())
```

```python
#df1.drop(columns=['N4W', inplace=True)
```

```python
df1
```

```python
train_df = df1[(df1['Date'] >= pd.Timestamp(datetime.date(2011, 2, 4))) & 
               (df1['Date'] <= pd.Timestamp(datetime.date(2011, 2, 4) + datetime.timedelta(weeks=71)))]
train_df
```

```python
test_df = df1[df1['Date'] > pd.Timestamp(datetime.date(2011, 2, 4) + datetime.timedelta(weeks=71))]
test_df
```

```python
train_df.dtypes
```

```python
X_train = train_df.drop(columns=['Store', 'Dept', 'Weekly_Sales', 'Date', 'Type', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Markdown', 'IsHoliday'])
X_test = test_df.drop(columns=['Store', 'Dept', 'Weekly_Sales', 'Date', 'Type', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Markdown', 'IsHoliday'])
y_train = train_df['Weekly_Sales']
y_test = test_df['Weekly_Sales']
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
cdf = pd.DataFrame(lm.coef_, X_train.columns, columns=['Coefficients'])
cdf
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
test_df.describe()
```

```python
line_df = test_df[['Date', 'Weekly_Sales']]
line_df['Actual/Prediction'] = 'Actual'
line_df1 = test_df[['Date']]
line_df1['Weekly_Sales'] = predictions
line_df1['Actual/Prediction'] = 'Prediction'
line_df = pd.concat([line_df, line_df1], ignore_index=True)
```

```python
line_df
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='Date', y='Weekly_Sales', data=line_df, hue='Actual/Prediction')
plt.tight_layout()
```
